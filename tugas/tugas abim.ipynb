```python
list_produk = [
    {
        'nama': 'Ayam',
        'stok': 100,
        'harga': 10000
    },
    {
        'nama': 'Daging',
        'stok': 100,
        'harga': 15000
    },
    {
        'nama': 'Anggur',
        'stok': 100,
        'harga': 20000
    }
]

cart = []

def konfirmasi_tambah():
    while True:
        user_input = input('\nKetik Y jika Anda yakin, N untuk batalkan: ').upper()

        if user_input == 'Y':
            break
        elif user_input == 'N':
            menambah_produk()
        else:
            user_input

def menampilkan_daftar_produk():
    print('\n------------- Daftar produk -------------')
    print('Index \t| Nama  \t| Stok\t| Harga')

    for index, item in enumerate(list_produk):
        print(f"{index} \t| {item['nama']}  \t| {item['stok']}\t| {item['harga']}")

def menambah_produk():
    while True:
        nama_produk = input('\nMasukkan nama produk: ').capitalize()

        while True:
            stok_produk = input('Masukkan stok produk: ')
            if stok_produk.isdigit() == False:
                print('Masukkan nilai yang sesuai')
                stok_produk
                continue
            else:
                stok_produk = int(stok_produk)

            while True:
                harga_produk = input('Masukkan harga produk: ')
                if harga_produk.isdigit() == False:
                    print('Masukkan nilai yang sesuai')
                    harga_produk
                    continue
                else:
                    harga_produk = int(harga_produk)
                    break
            break
        break

    konfirmasi_tambah()

    for item in list_produk:
        if (item['nama'] == nama_produk) and (item['harga'] == harga_produk):
            item['stok'] += stok_produk
            break
    else:
        list_produk.append({
            'nama': nama_produk,
            'stok': stok_produk,
            'harga': harga_produk
        })

    menampilkan_daftar_produk()
    main_menu()

def menghapus_produk():
    menampilkan_daftar_produk()

    while True:
        index_produk = input('Masukkan index produk yang ingin dihapus: ')
        if index_produk.isdigit() == False:
            print('Masukkan angka index yang sesuai')
            index_produk
        else:
            index_produk = int(index_produk)
            if index_produk not in range(len(list_produk)):
                print(f'Tidak ada index {index_produk} pada database. Silakan masukkan angka index yang sesuai.')
            else:
                print('\nHapus data produk dari database? (Y/N)')

                while True:
                    user_input = input('\nKetik Y jika Anda yakin, N untuk batalkan: ').upper()
                    if user_input == 'Y':
                        break
                    elif user_input == 'N':
                        main_menu()
                    else:
                        print('Input tidak sesuai. Silakan ketik Y atau N untuk konfirmasi')
                        user_input

                break

    del list_produk[index_produk]

    menampilkan_daftar_produk()
    main_menu()


def membeli_produk():
    menampilkan_daftar_produk()

    while True:
        index_produk = int(input('Masukkan index produk yang ingin dibeli: '))
        jumlah_produk = int(input('Masukkan jumlah yang ingin dibeli: '))

        if jumlah_produk > list_produk[index_produk]['stok']:
            print(f"Stok tidak cukup. Stok {list_produk[index_produk]['nama']} tersisa {list_produk[index_produk]['stok']} produk")
        else:
            cart.append({
                'nama': list_produk[index_produk]['nama'],
                'jumlah': jumlah_produk,
                'harga': list_produk[index_produk]['harga'],
                'index': index_produk
            })

        for item in cart:
            list_produk[item['index']]['stok'] -= item['jumlah']

        print('Isi Cart:')
        print(f'Nama\t| Jumlah\t| Harga')

        for item in cart:
            print(f"{item['nama']}\t| {item['jumlah']}\t\t| {item['harga']}")

        break

    while True:
        checker = input('Mau beli yang lain? (Ya/Tidak) = ').lower()

        if(checker == 'ya'):
            index_produk = int(input('Masukkan index produk yang ingin dibeli: '))
            print(f"Sisa stok {list_produk[index_produk]['nama']} adalah {list_produk[index_produk]['stok']}.")

            if(list_produk[index_produk]['stok'] == 0):
                print('Maaf, stok produk yang dipilih sedang kosong. Silakan pilih produk yang lain jika mau.')
                continue
            else:
                jumlah_produk_tambahan = int(input('Masukkan jumlah yang ingin dibeli: '))
                cart.append([list_produk[index_produk]['nama'], jumlah_produk_tambahan, list_produk[index_produk]['harga'], index_produk])
                list_produk[index_produk]['stok'] -= jumlah_produk_tambahan

                if(jumlah_produk_tambahan > list_produk[index_produk]['stok']):
                    print(f'Stok tidak cukup, masukkan jumlah yang sesuai dengan sisa stok.')
                else:
                    cart.append([list_produk[index_produk]['nama'], jumlah_produk_tambahan, list_produk[index_produk]['harga'], index_produk])
                    list_produk[index_produk]['stok'] -= jumlah_produk_tambahan

        else:
            break

    print('------------- Daftar Belanja -------------')
    print('Nama  \t| Jumlah\t| Harga \t| Total Harga')

    total_harga = 0

    for item in cart:
        print(f"{item['nama']}\t| {item['jumlah']}\t\t| {item['harga']} \t|  {item['jumlah'] * item['harga']}")
        total_harga += item['jumlah'] * item['harga']
        break

    while True:
        print(f'Total yang harus dibayar = {total_harga}')
        if total_harga != 0:
            jumlah_uang = int(input('Masukkan jumlah uang: '))

            if(jumlah_uang > total_harga):
                kembali = jumlah_uang - total_harga
                print(f'Terima kasih \n\nUang kembalian Anda: {kembali}')
                if total_harga > 500000:
                    diskon = 0.1 * total_harga
                    print(f"Anda mendapatkan diskon 10% sebesar {diskon}")
                    total_harga -= diskon
                elif total_harga > 300000:
                    diskon = 0.08 * total_harga
                    print(f"Anda mendapatkan diskon 8% sebesar {diskon}")
                    total_harga -= diskon
                elif total_harga > 200000:
                    diskon = 0.05 * total_harga
                    print(f"Anda mendapatkan diskon 5% sebesar {diskon}")
                    total_harga -= diskon
                print(f'Total harga setelah diskon = {total_harga}')
                cart.clear()
                break

            elif(jumlah_uang == total_harga):
                print('Terima kasih')
                cart.clear()
                break

            else:
                kekurangan = total_harga - jumlah_uang
                print(f'Uang Anda kurang sebesar {kekurangan}')

        else:
            print(f'Terima kasih telah berkunjung.')
            break

    main_menu()


def main_menu():
    while True:
        pilih_menu = input ('''
            Selamat Datang di Toko Abimas

            1. Daftar produk
            2. Menambah produk
            3. Menghapus produk
            4. Membeli produk
            5. Keluar program

            Silakan pilih nomor menu: ''')

        print()
        if pilih_menu == '1':
            menampilkan_daftar_produk()
        elif pilih_menu == '2':
            menampilkan_daftar_produk()
            menambah_produk()
        elif pilih_menu == '3':
            menghapus_produk()
        elif pilih_menu == '4':
            membeli_produk()
        elif pilih_menu == '5':
            break

main_menu()
```
