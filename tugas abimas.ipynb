```python
list_buah = [
    {
        'nama': 'Apel',
        'stok': 100,
        'harga': 10000
    },
    {
        'nama': 'Jeruk',
        'stok': 100,
        'harga': 15000
    },
    {
        'nama': 'Anggur',
        'stok': 100,
        'harga': 20000
    }
]

cart = []

def konfirmasi_tambah():
    while True:
        user_input = input('\nKetik Y jika Anda yakin, N untuk batalkan: ').upper()

        if user_input == 'Y':
            break
        elif user_input == 'N':
            menambah_buah()
        else:
            user_input

def menampilkan_daftar_buah():
    print('\n------------- Daftar Buah -------------')
    print('Index \t| Nama  \t| Stok\t| Harga')

    for index, item in enumerate(list_buah):
        print(f"{index} \t| {item['nama']}  \t| {item['stok']}\t| {item['harga']}")

def menambah_buah():
    while True:
        nama_buah = input('\nMasukkan nama buah: ').capitalize()

        while True:
            stok_buah = input('Masukkan stok buah: ')
            if stok_buah.isdigit() == False:
                print('Masukkan nilai yang sesuai')
                stok_buah
                continue
            else:
                stok_buah = int(stok_buah)

            while True:
                harga_buah = input('Masukkan harga buah: ')
                if harga_buah.isdigit() == False:
                    print('Masukkan nilai yang sesuai')
                    harga_buah
                    continue
                else:
                    harga_buah = int(harga_buah)
                    break
            break
        break

    konfirmasi_tambah()

    for item in list_buah:
        if (item['nama'] == nama_buah) and (item['harga'] == harga_buah):
            item['stok'] += stok_buah
            break
    else:
        list_buah.append({
            'nama': nama_buah,
            'stok': stok_buah,
            'harga': harga_buah
        })

    menampilkan_daftar_buah()
    main_menu()

def menghapus_buah():
    menampilkan_daftar_buah()

    while True:
        index_buah = input('Masukkan index buah yang ingin dihapus: ')
        if index_buah.isdigit() == False:
            print('Masukkan angka index yang sesuai')
            index_buah
        else:
            index_buah = int(index_buah)
            if index_buah not in range(len(list_buah)):
                print(f'Tidak ada index {index_buah} pada database. Silakan masukkan angka index yang sesuai.')
            else:
                print('\nHapus data buah dari database? (Y/N)')

                while True:
                    user_input = input('\nKetik Y jika Anda yakin, N untuk batalkan: ').upper()
                    if user_input == 'Y':
                        break
                    elif user_input == 'N':
                        main_menu()
                    else:
                        print('Input tidak sesuai. Silakan ketik Y atau N untuk konfirmasi')
                        user_input

                break

    del list_buah[index_buah]

    menampilkan_daftar_buah()
    main_menu()


def membeli_buah():
    menampilkan_daftar_buah()

    while True:
        index_buah = int(input('Masukkan index buah yang ingin dibeli: '))
        jumlah_buah = int(input('Masukkan jumlah yang ingin dibeli: '))

        if jumlah_buah > list_buah[index_buah]['stok']:
            print(f"Stok tidak cukup. Stok {list_buah[index_buah]['nama']} tersisa {list_buah[index_buah]['stok']} buah")
        else:
            cart.append({
                'nama': list_buah[index_buah]['nama'],
                'jumlah': jumlah_buah,
                'harga': list_buah[index_buah]['harga'],
                'index': index_buah
            })

        for item in cart:
            list_buah[item['index']]['stok'] -= item['jumlah']

        print('Isi Cart:')
        print(f'Nama\t| Jumlah\t| Harga')

        for item in cart:
            print(f"{item['nama']}\t| {item['jumlah']}\t\t| {item['harga']}")

        break

    while True:
        checker = input('Mau beli yang lain? (Ya/Tidak) = ').lower()

        if(checker == 'ya'):
            index_buah = int(input('Masukkan index buah yang ingin dibeli: '))
            print(f"Sisa stok {list_buah[index_buah]['nama']} adalah {list_buah[index_buah]['stok']}.")

            if(list_buah[index_buah]['stok'] == 0):
                print('Maaf, stok buah yang dipilih sedang kosong. Silakan pilih buah yang lain jika mau.')
                continue
            else:
                jumlah_buah_tambahan = int(input('Masukkan jumlah yang ingin dibeli: '))
                cart.append([list_buah[index_buah]['nama'], jumlah_buah_tambahan, list_buah[index_buah]['harga'], index_buah])
                list_buah[index_buah]['stok'] -= jumlah_buah_tambahan

                if(jumlah_buah_tambahan > list_buah[index_buah]['stok']):
                    print(f'Stok tidak cukup, masukkan jumlah yang sesuai dengan sisa stok.')
                else:
                    cart.append([list_buah[index_buah]['nama'], jumlah_buah_tambahan, list_buah[index_buah]['harga'], index_buah])
                    list_buah[index_buah]['stok'] -= jumlah_buah_tambahan

        else:
            break

    print('------------- Daftar Belanja -------------')
    print('Nama  \t| Jumlah\t| Harga \t| Total Harga')

    total_harga = 0

    for item in cart:
        print(f"{item['nama']}\t| {item['jumlah']}\t\t| {item['harga']} \t|  {item['jumlah'] * item['harga']}")
        total_harga += item['jumlah'] * item['harga']
        break

    while True:
        print(f'Total yang harus dibayar = {total_harga}')
        if total_harga != 0:
            jumlah_uang = int(input('Masukkan jumlah uang: '))

            if(jumlah_uang > total_harga):
                kembali = jumlah_uang - total_harga
                print(f'Terima kasih \n\nUang kembalian Anda: {kembali}')
                if total_harga > 500000:
                    diskon = 0.1 * total_harga
                    print(f"Anda mendapatkan diskon 10% sebesar {diskon}")
                    total_harga -= diskon
                elif total_harga > 300000:
                    diskon = 0.08 * total_harga
                    print(f"Anda mendapatkan diskon 8% sebesar {diskon}")
                    total_harga -= diskon
                elif total_harga > 200000:
                    diskon = 0.05 * total_harga
                    print(f"Anda mendapatkan diskon 5% sebesar {diskon}")
                    total_harga -= diskon
                print(f'Total harga setelah diskon = {total_harga}')
                cart.clear()
                break

            elif(jumlah_uang == total_harga):
                print('Terima kasih')
                cart.clear()
                break

            else:
                kekurangan = total_harga - jumlah_uang
                print(f'Uang Anda kurang sebesar {kekurangan}')

        else:
            print(f'Terima kasih telah berkunjung.')
            break

    main_menu()


def main_menu():
    while True:
        pilih_menu = input('''
            Selamat Datang di Pasar Buah

            List Menu:
            1. Menampilkan daftar buah
            2. Menambah buah
            3. Menghapus buah
            4. Membeli buah
            5. Keluar program

            Silakan pilih nomor menu: ''')

        print()
        if pilih_menu == '1':
            menampilkan_daftar_buah()
        elif pilih_menu == '2':
            menambah_buah()
        elif pilih_menu == '3':
            menghapus_buah()
        elif pilih_menu == '4':
            membeli_buah()
        elif pilih_menu == '5':
            break

main_menu()
```
