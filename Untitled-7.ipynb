```python
items = [{'name': 'ayam', 'quantity': 10, 'price': 3000}, {'name': 'telur', 'quantity': 10, 'price': 2000}]
while True:
    print('================ Toko Abimas ================')
    print('1. List Produk\n2. Pembelian Produk\n3. Penambahan Produk')
    choice = input('Pilih opsi : ')

    if choice == '1' :
        print('------------------List Produk------------------')
        print('Jumlah jenis produk : ',len(items))
        while len(items) != 0:
            print('List Produk yang Tersedia')
            print(items)
            break

    elif choice == '2' :
        print('------------------Pembelian Produk------------------')
        print(items)
        purchase_item = input('Masukkan Produk yang dibeli : ')
        quantity_item = int(input('Masukkan jumlah :'))
        for item in items:
            if purchase_item.lower() == item['name'].lower() :
                if item['quantity'] != 0  :
                    print('Pembayaran ', (item['price']*quantity_item) , 'di kasir.')
                    item['quantity'] -= 1
                else: 
                    print('Produk kosong.')       

    elif choice == '3' :
        print('------------------Penambahan Produk------------------')
        print('Ikuti instruksi untuk penambahan produk')
       
        item = {}
        item['name'] = input('Nama produk : ')
        while True:
            try:
                item['quantity'] = int(input('Item quantity : '))
                break
            except ValueError:
                print('Satuan berupa angka')
        while True:
            try:
                item['price'] = int(input('Rp : '))
                break
            except ValueError:
                print('Satuan berupa nominal')
        print('Produk berhasil ditambahkan.')
        items.append(item)

    elif choice == '4' :
        print('------------------Penghapusan Produk------------------')
        print('Ikuti instruksi untuk penghapusan produk')
       
        item = {}
        item['name'] = input('Nama produk : ')
        item['quantity'] = int(input('Item quantity : '))
        item['price'] = int(input('Rp : '))

        if item in items :
            print ('Produk berhasil dihapus')
            items.remove(item)
        
        elif item not in items :
            print ('Produk tidak ada dalam list')
            break

    elif choice == '6':
        print ('------------------Terima Kasih------------------')
        break
    

    
```
